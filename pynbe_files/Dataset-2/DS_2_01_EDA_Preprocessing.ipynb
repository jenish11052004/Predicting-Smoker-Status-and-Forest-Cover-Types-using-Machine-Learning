```python
plt.figure(figsize=(12, 10))
correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Check for high correlation (> 0.8)
high_corr_pairs = []
corr_matrix = df[numerical_cols].corr().abs()

# Iterate over the correlation matrix to find high correlations
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] > 0.8:
            colname_i = corr_matrix.columns[i]
            colname_j = corr_matrix.columns[j]
            high_corr_pairs.append((colname_i, colname_j, corr_matrix.iloc[i, j]))

if high_corr_pairs:
    print("Highly correlated features (Correlation > 0.8):")
    for pair in high_corr_pairs:
        print(f"{pair[0]} and {pair[1]}: {pair[2]:.2f}")
    print("\nConsider removing one feature from each pair to reduce multicollinearity.")
else:
    print("No highly correlated features found (Correlation > 0.8).")
```

## Preprocessing


### 1. Outlier Removal using Modified Z-score Method


```python
# -----------------------------
# Outlier Removal using Modified Z-Score
# -----------------------------

import numpy as np
import pandas as pd

# Keep a copy of original data
df_original = df.copy()

# Identify numerical continuous features
numerical_cols = df_original.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Remove soil types + wilderness (0/1 dummy variables)
categorical_like = [col for col in numerical_cols if 
                    'Soil_Type' in col or 'Wilderness' in col or 'Cover_Type' in col]

numerical_cols = [col for col in numerical_cols if col not in categorical_like]

print("Numerical columns used for outlier detection:")
print(numerical_cols)

# Modified Z-score function
def modified_z_score(series):
    median = series.median()
    mad = np.median(np.abs(series - median))
    if mad == 0:
        # If MAD = 0, no variation  no outliers
        return np.zeros(len(series))
    return 0.6745 * (series - median) / mad

# Compute modified z-scores for each numerical column
z_scores = pd.DataFrame({
    col: modified_z_score(df_original[col]) 
    for col in numerical_cols
})

# Identify outliers
outlier_mask = (np.abs(z_scores) > 3.5).any(axis=1)

# Remove outliers
df_removed = df_original[outlier_mask]     # only outliers
df_clean   = df_original[~outlier_mask]    # cleaned dataset

# Update df for further processing
df = df_clean.copy()

print("\nOutlier removal results:")
print(f"Original shape: {df_original.shape}")
print(f"New shape: {df_clean.shape}")
print(f"Rows removed: {df_removed.shape[0]}")
```

```python
# -----------------------------
# Inspect Removed Outliers (Severity Check)
# -----------------------------
from scipy.stats import zscore

# Re-identify numeric columns (matching earlier logic)
numeric_cols = df_original.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Remove dummy categorical-like columns
categorical_like = [col for col in numeric_cols 
                    if 'Soil_Type' in col or 'Wilderness' in col]

numeric_cols = [col for col in numeric_cols if col not in categorical_like]

# Compute standard z-score for reference (this is NOT used for removal)
z_all = df_original[numeric_cols].apply(zscore)

# Extract z-scores ONLY for removed outliers
df_removed_z = z_all.loc[df_removed.index]

print("\nZ-score summary of removed rows:")
print(df_removed_z.describe())
```

### 2. Remove Skewness

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
df[numerical_cols] = pt.fit_transform(df[numerical_cols])
```

```python
from sklearn.preprocessing import RobustScaler

df_scaled = df.copy()

scaler = RobustScaler()

# Fit and transform only the numerical columns
df_scaled[numerical_cols] = scaler.fit_transform(df_scaled[numerical_cols])


print("\nAfter scaling:")
print(df_scaled[numerical_cols].describe())

df = df_scaled.copy()

```
